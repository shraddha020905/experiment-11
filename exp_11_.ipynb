```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    "Name": ["A", "B", "C", "D"],
    "Marks": [85, 90, 78, 88]
}
df = pd.DataFrame(data)
df
```

```python
df.head()
df.describe()
df["Marks"].mean()
```

```python
plt.bar(df["Name"], df["Marks"])
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```
